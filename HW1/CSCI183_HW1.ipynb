```python
# Question 1: a Python program to print the number and sum of prime numbers which are less than or equal to a given number.

def is_prime(x):
    if x==0 or x==1:
        return False
    for i in range(2, x):
        if ((x%i)==0):
            return False
    return True
# note 1 is not a prime.

def printNum(x):
    primeSum = 0
    for i in range(x+1):
        print(i)
        if is_prime(i):
            primeSum += i
    print(f"the sum of all primes less than or equal to {i} is {primeSum}")

# Testing Example:
printNum(4) # should result 2+3=5
printNum(12) # should result 2+3+5+7+11=28
```

```python
# Q2: Write a Python program to find the largest product of the pair of adjacent elements from a given list of integers.
def maxProduct(l):
  max_val = float('-inf')
  maxIndex = 0
  for i in range(len(l)-1):
    product = l[i] * l[i+1]
    if product > max_val:
      max_val = product
      maxIndex = i
  print(f"the largest adjacent product is {max_val} from number indexing {maxIndex} and {maxIndex+1}")


# Testing Example
maxProduct([3, 6, -2, -5, 7, 3])  # Example list, largest should be 7*3=21
```

```python
# Q3. Given a password, we have to categorize it as valid or invalid. There are some checks that need to be met to be a valid password. For a invalid password, we need to return the reason for it to be invalid. [10 points]

# Conditions for a valid password are:

# Should have at least one number.
# Should have at least one uppercase and one lowercase character.
# Should have at least one special symbol.
# Should be between 6 to 20 characters long.



def isValid(passwd):
  # check condition 1: Should have at least one number.
  if re.search('\d+', passwd) == None:
    print("Failed condition 1 for valid password: Should have at least one number.")
    return False

  # check condition 2: Should have at least one uppercase and one lowercase character.
  if re.search('[a-z]+', passwd) == None or re.search('[A-Z]+', passwd) == None:
    print("Failed condition 2 for valid password: Should have at least one uppercase and one lowercase character.")
    return False

  # check condition 3: Should have at least one special symbol.
  if re.search('[~`!@#$%^&*()\-_+={}[\]|\\;:"<>,./?]', passwd) is None:
    print("Failed condition 3 for valid password: Should have at least one special symbol")
    return False

  # check condition 4: Should be between 6 to 20 characters long.
  if len(passwd)<6 or len(passwd)>20:
    print("Failed condition 4 for valid password: Should be between 6 to 20 characters long.")
    return False

  # valid password
  print("Is valid password, four conditions matched")
  return True


# Testing Example
isValid('Password!!')
isValid('1234567890!!')
isValid('Password1')
isValid('Pas1+')
isValid('Password123!!')
print("\n")
```

```python
# Q4. Given a password, we have to categorize it as a strong or weak one. There are some checks that need to be met to be a strong password. For a weak password, we need to return the reason for it to be weak. [10 points]

# Conditions to be fulfilled for strong password are:

# Minimum 9 characters and maximum 20 characters.
# Cannot be a newline or a space
# There should not be three or more repeating characters in a row.
# The same string pattern(minimum of two character length) should not be repeated.

def isStrong(passwd):
  #check if is a valid password
  if not isValid(passwd):
    print("Not a valid password, unable to test strong or weak.\n")
    return False

  # check condition 1: Minimum 9 characters and maximum 20 characters.
  if len(passwd)<9 or len(passwd)>20:
    print("Failed condition 1 for strong password: Minimum 9 characters and maximum 20 characters.\n")
    return False

  # check condition 2: Cannot be a newline or a space
  if re.search('[\n\s]', passwd) != None:
    print("Failed condition 2 for strong password: Cannot be a newline or a space.\n")
    return False

  # check condition 3: There should not be three or more repeating characters in a row.
  if re.search(r'(.)\1{2}', passwd) != None:
    print("Failed condition 3 for strong password: There should not be three or more repeating characters in a row.\n")
    return False

  #check condition 4: The same string pattern(minimum of two character length) should not be repeated.
  for size in range(2, len(passwd)//2):
    for i in range(len(passwd)-size+1):
      if passwd[i:(i+size)] in passwd[(i+size):]:
        print("Failed condition 4 for strong password: The same string pattern(minimum of two character length) should not be repeated.\n")
        return False

  print("This is a strong password, four conditions matched.\n")
  return True


# Testing Example
isStrong("Password!!")
isStrong("Pa0cdi!!")
isStrong("Pa0c cd1i!!")
isStrong("Passsword12!")
isStrong("Password1212!")
isStrong("Password12!")
print("\n")
```

```python
# Q5. Write a Python program to convert a Python DataFrame to a dictionary object.  [5 points]

# Contents of the DataFrame:

#     C1  C2  C3

# R1   1   2   3

# R2   4   5   6

# R3   7   8   9

# DataFrame as a dictionary:

# {'C1': {'R1': 1, 'R2': 4, 'R3': 7}, 'C2': {'R1': 2, 'R2': 5, 'R3': 8}, 'C3': {'R1': 3, 'R2': 6, 'R3': 9}}


def df_to_dict(df):
    keys = df.columns
    keys_keys = df.index
                
    df_dict = {}    
    for column_name in keys:
        
        each_column = {}
        for row_name in keys_keys:
            each_column[row_name] = df.loc[row_name, column_name]
        df_dict[column_name] = each_column
    return df_dict


# testing example
contents = [['C1', 'C2', 'C3'], [1, 2, 3], [4, 5, 6], [7, 8, 9]]
indexs = ['R1','R2','R3']
df1 = pd.DataFrame(contents[1:], columns=contents[0], index = indexs)

dicts1 = df_to_dict(df1)
display(dicts1)
```

```python
# Q1. Write a Python program that matches a word containing 'z', not at the start or end of the word.

def find_z(word):
    match = re.match('[a-yA-Y]+z[a-yA-Y]+', word)
    return match

    
# testing example
word = "qefiuhweifjweoufhwefdzhew9jzqeiufwnefujfweo"
print(find_z(word))


```

```python
# Q2. You are given a list of courses of a student: [5 points]
# ["CSCI 10", "MATH 9", "MATH 11", "CSCI 60", "CSCI 182"]
# Write a python program that helps you to find the CSCI courses that the student has taken. 


courselist = ["csci 10", "MATH 9", "MATH 11", "Csci 60", "CSCI 182"]
csci_course = []

for course in courselist:
    course = course.strip() # in case there's any extra space
    if re.match('[Cc][Ss][Cc][Ii]', course):
        csci_course.append(course)
        print(course)

print(csci_course)
```
